```python
# state = 'qjdisidjq/9/1b5b1/o1o1o1o1o/9/9/O1O1O1O1O/1B5B1/9/QJDISIDJQ'
state  = 'qjdisidjq/9/1b5b1/o1o1o1o1o/9/9/O1O1O1O1O/1B5B1/9/QJDISIDJQ'
BOARD_WIDTH = 9; BOARD_HEIGHT=10
state_to_board_dict = {
    'k': 'n',
    'K': 'N',
    'e': 'b',
    'E': 'B',
    'm': 'a',
    'M': 'A',
    's': 'k',
    'S': 'K',
    'r': 'r',
    'R': 'R',
    'p': 'p',
    'P': 'P',
    'c': 'c',
    'C': 'C',
}

replace_dict = {
    'n': 'k',
    'N': 'K',
    'b': 'e',
    'B': 'E',
    'a': 'm',
    'A': 'M',
    'k': 's',
    'K': 'S',
    'r': 'r',
    'R': 'R',
    'p': 'p',
    'P': 'P',
    'c': 'c',
    'C': 'C',
}

def state_to_board(state):
    board = [['.' for col in range(9)] for row in range(10)]
    x = 0
    y = 9
    for k in range(0, len(state)):
        ch = state[k]
        if ch == ' ':
            break
        if ch == '/':
            x = 0
            y -= 1
        elif ch >= '1' and ch <= '9':
            for i in range(int(ch)):
                board[y][x] = '.'
                x = x + 1
        else:
            board[y][x] = swapcase(ch, s2b=True)
            x = x + 1
    return board

def swapcase(a, s2b=False):
    if a.isalpha():
        if s2b:
            a = state_to_board_dict[a]
        else:
            a = replace_dict[a]
        return a.lower() if a.isupper() else a.upper()
    return a

def step(state, action):
    board = state_to_board(state)
    print(board, '\n')
    if board[int(action[1])][int(action[0])] == '.':
        raise ValueError(f"No chessman in {action}, state = {state}")
    board[int(action[3])][int(action[2])] = board[int(action[1])][int(action[0])]
    board[int(action[1])][int(action[0])] = '.'
    state = board_to_state(board)
    print(state)
    return fliped_state(state)

def board_to_state(board):
    c = 0
    fen = ''
    for i in range(10 - 1, -1, -1):
        c = 0
        for j in range(9):
            if board[i][j] == '.':
                c = c + 1
            else:
                if c > 0:
                    fen = fen + str(c)
                fen = fen + swapcase(board[i][j])
                c = 0
        if c > 0:
            fen = fen + str(c)
        if i > 0:
            fen = fen + '/'
    return fen

def fliped_state(state):
    rows = state.split('/')
    def swapcase(a):
        if a.isalpha():
            return a.lower() if a.isupper() else a.upper()
        return a
    def swapall(aa):
        return "".join([swapcase(a) for a in aa])

    return "/".join([swapall(reversed(row)) for row in reversed(rows)])

def init_state(INIT_STATE):
    board = [['.'] * BOARD_WIDTH for _ in range(BOARD_HEIGHT)]
    x = y = 0
    for ch in INIT_STATE:
        if ch == '/':
            x += 1
            y = 0
        elif '1' <= ch <= '9':
            y += int(ch)
        else:
            board[x][y] = ch
            y += 1
    return board

init_state(state)
```
